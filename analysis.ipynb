```python
import pandas as pd
import matplotlib.pyplot as plt

# Define file paths
lecture_block_path = "iiitd/Lecture Block/Building Total Mains/0/Energy.csv"
academic_block_path = "iiitd/Academic Block/Building Total Mains/0/Energy.csv"

# Load CSV files
lecture_df = pd.read_csv(lecture_block_path)
academic_df = pd.read_csv(academic_block_path)

# Naming columns to understand what's happening
lecture_df.columns = ["Timestamp", "Energy_Consumption"]
academic_df.columns = ["Timestamp", "Energy_Consumption"]

# Convert timestamp to datetime format
lecture_df["Timestamp"] = pd.to_datetime(lecture_df["Timestamp"], unit='ms')
academic_df["Timestamp"] = pd.to_datetime(academic_df["Timestamp"], unit='ms')

# Plot energy consumption
plt.figure(figsize=(12, 6))
plt.plot(lecture_df["Timestamp"], lecture_df["Energy_Consumption"], label="Lecture Block", alpha=0.7)
plt.plot(academic_df["Timestamp"], academic_df["Energy_Consumption"], label="Academic Block", alpha=0.7)

# Formatting
plt.xlabel("Time")
plt.ylabel("Energy Consumption (Unknowkn Units)")
plt.title("Energy Consumption Over Time")
plt.legend()
plt.xticks(rotation=45)
plt.grid()
plt.show()

# Print total energy consumption
print("Total Energy Consumption:")
print(f"Lecture Block: {lecture_df['Energy_Consumption'].sum():,.2f}")
print(f"Academic Block: {academic_df['Energy_Consumption'].sum():,.2f}")
```
